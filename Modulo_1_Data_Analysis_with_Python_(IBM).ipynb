**IMPORTANDO Y EXPORTANDO DATA EN PYTHON**
Importando: Se refiere al proceso de carga y lectura de datos desde varios recursos. Se tienen dos propiedades importantes:
1. El formato: Común es .csv, .json, .xlsx, etc
2. La ruta del conjunto de datos: puede ser local (carperta en el computador o desde internet)

```python
# Importando y exportando data en Python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url) # Importa los datos con encabezados
df = pd.read_csv(url, header=None) # Importa los datos sin encabezados
```

**Printing the dataframe in Python**
1. df imprime el dataframe completo (no se recomienda para datasets grandes)
2. df.head(n) sirve para mostrar las primeras n filas del dataframe
3. df.tail(n) muestra las últimas n filas del dataframe
