**Analysis of Variance (ANOVA)**

Para comparación estadística de grupos, por ejemplo, promediar precio de carros de diferentes marcas.

Este método sirve para buscar la correlación entre diferentes grupos de variables categóricas. Se puede obtener la variación entre dos grupos de variables (F-test score) o el grado de confianza (p-value).

En el siguiente ejemplo se puede ver la aplicabilidad del método:

Se compara la variable categórica Honda contra Subaru y Jaguar, los precios de Honda frente a Subaru están en el mismo orden por lo que el F-test score es menor que 1, pero cuando se compara Honda con Jaguar, la diferencia en precio es mucho mayor y por tal razón el F-score es 401 aprox.
